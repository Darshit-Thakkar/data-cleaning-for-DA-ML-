```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
car_df = pd.read_csv("car_price.csv")
```

```python
car_df.info()
```

```python
# remove the unnecessary columns
car_df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
car_df.info()
```

```python
# seprate the car company name 

def com_name(x):
    return x[:x.index(" ")]

car_df["company_name"] = car_df["car_name"].apply(com_name)
```

```python
# seprate the car name 

def car_name(x):
    return x[x.index(" ")+1:]

car_df["car_name"] = car_df["car_name"].apply(car_name)
```

```python
car_df.head(5)
```

```python
# remove the coma (,) from the car_prices_in_rupee column 
car_df["car_prices_in_rupee"]=car_df["car_prices_in_rupee"].str.replace(",","")
```

```python
# build the function that replace the lakh and crore string into the numeric value (integer) 

def rupe(x):
    p = x.split(" ")
    try:
        if p[1]=="Lakh":
            return str(round(float(p[0])*100000,1))
        elif p[1]=="Crore":
            return str(round(float(p[0])*10000000,1))
    except:
        return x
    
car_df["car_prices_in_rupee"]=car_df["car_prices_in_rupee"].apply(rupe).astype("float")
# car_df["car_prices_in_rupee"].astype("float")
```

```python
car_df.info()
```

```python
# replace kms string into the numeric value (integer) 

car_df["kms_driven"] = car_df["kms_driven"].str.replace(",","")
car_df["kms_driven"] = car_df["kms_driven"].str.replace(" kms","")
car_df["kms_driven"] = car_df["kms_driven"].astype("int64")
```

```python
# remove cc from the engine column 

car_df["engine"] = car_df["engine"].str.replace(" cc","")
car_df["engine"] = car_df["engine"].astype("int64")
```

```python
# remove seats from the seats column 

car_df["Seats"] = car_df["Seats"].str.replace(" Seats","")
car_df["Seats"] = car_df["Seats"].astype("int64")
```

```python
car_df.info()
```

# find the outlier of engine 

```python
q1 = np.quantile(car_df['engine'],0.25)
q3 = np.quantile(car_df['engine'],0.75)
iqr = q3 - q1
```

```python
# finding the min and max outlier 
min_r = q1 - (1.5*iqr)
max_r = q3 + (1.5*iqr)
min_r , max_r
```

```python
# show the max 

car_df = car_df[car_df['engine']<=max_r]
car_df
```

# scaling the data 

```python
ss = StandardScaler()
new_car_df = pd.DataFrame(ss.fit_transform(car_df),columns=car_df.columns)
new_car_df.head(5)
```
